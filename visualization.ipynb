Based on rotten tomato 

```python
import pandas as pd
import matplotlib.pyplot as plt

df_rotten = pd.read_sql(rotten_tomatoes_query, conn)

plt.figure(figsize=(12, 6))
plt.hist(df_rotten[df_rotten['Netflix'] == 1]['RottenTomatoesScore'], alpha=0.5, label='Netflix')
plt.hist(df_rotten[df_rotten['Hulu'] == 1]['RottenTomatoesScore'], alpha=0.5, label='Hulu')
plt.hist(df_rotten[df_rotten['Prime Video'] == 1]['RottenTomatoesScore'], alpha=0.5, label='Prime Video')
plt.hist(df_rotten[df_rotten['Disney+'] == 1]['RottenTomatoesScore'], alpha=0.5, label='Disney+')
plt.xlabel('Rotten Tomatoes Score')
plt.ylabel('Number of Movies')
plt.title('Distribution of Rotten Tomatoes Scores Across Streaming Platforms')
plt.legend()
plt.show()
```

```python


df_avg_rating = pd.read_sql(average_rating_query, conn)

data = [df_avg_rating[df_avg_rating['Netflix'] == 1]['averageRating'],
        df_avg_rating[df_avg_rating['Hulu'] == 1]['averageRating'],
        df_avg_rating[df_avg_rating['Prime Video'] == 1]['averageRating'],
        df_avg_rating[df_avg_rating['Disney+'] == 1]['averageRating']]
labels = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']

sns.set(style="whitegrid", palette="pastel")

plt.figure(figsize=(10, 6))
sns.boxplot(data=data, width=0.5)

plt.xticks(range(len(labels)), labels)
plt.ylabel('Average Rating')
plt.title('Distribution of Average Ratings Across Streaming Platforms')

palette = sns.color_palette("coolwarm", len(labels))
for i, artist in enumerate(plt.gca().artists):
    # Set the linecolor on the artist to the facecolor, and set the facecolor to None
    col = palette[i]
    artist.set_edgecolor(col)
    for j in range(i*6, i*6+6):
        line = plt.gca().lines[j]
        line.set_color(col)
        line.set_mfc(col)
        line.set_mec(col)

plt.show()
```
